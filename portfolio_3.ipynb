```python
# Det valgte dataset bliver importeret, med søge query "skitser"
# Det bliver lagt ind i et dataframe
import requests
import pandas as pd
from pandas.io.json import json_normalize

smk_search_url = 'https://api.smk.dk/api/v1/art/search/?keys=skitser&offset=0&rows=150'

params = {
    'q': 'skitser', 
    'items': 'items',
    'encoding': 'json',
    'n': 150
}

response = requests.get(smk_search_url, params=params)
print(response)
json = response.json()
df = json_normalize(json['items'])
```

```python
df.head()
```

```python
def flatten_json(y):
    out = {}

    def flatten(x, name=''):
        if type(x) is dict:
            for a in x:
                flatten(x[a], name + a + '_')
        elif type(x) is list:
            i = 0
            for a in x:
                flatten(a, name + str(i) + '_')
                i += 1
        else:
            out[name[:-1]] = x

    flatten(y)
    return out
```

```python
dic_flattened = [flatten_json(d) for d in json['items']]
df= json_normalize(dic_flattened)
```

```python
# denne funktion tæller via pandas value_counts() hvor mange poster har hvilken værdi.
# I dette tilfælde, om posten er PublicDomain (=True), eller ikke (=False)
# derefter udregnes der en procent del af det totale udtræk
publicdomain = df['public_domain'].value_counts()
publicdomainprocent = df['public_domain'].value_counts(normalize=True).mul(100).round(1).astype(str) + '%'
domaindf = pd.DataFrame(columns=['Antal', 'Procent'], index=['Public Domain', 'Ikke Public',], data=zip(publicdomain,publicdomainprocent))
print(domaindf)
```
